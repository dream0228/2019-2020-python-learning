```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

## contour練習-畫等高線圖

```python
step = 0.01      #每隔0.01取一個點
x = np.arange(-10,10,step)
y = np.arange(-10,10,step)
```

```python
X,Y = np.meshgrid(x,y)
```

```python
Z = 3*X**2+13*Y**2-32
```

```python
plt.figure(figsize=(10,8))   #設置圖的大小，長10，寬8
plt.contourf(X,Y,Z) 
```

```python
contour = plt.contour(X,Y,Z,[50,100,150,200],colors='b')     #畫50,100,150,200的線，顏色為藍色
plt.clabel(contour,fontsize=10,colors=('b','r','g','y'))     #標示出各值，設定字體大小10，颜色是黑、紅、綠、黃
```

```python
cset = plt.contourf(X,Y,Z,6,cmap=plt.cm.Wistia) 
contour = plt.contour(X,Y,Z,[50,100,150,200],colors='k')     
plt.clabel(contour,fontsize=10,colors=('b','r','g','y'))
plt.colorbar(cset)
```
