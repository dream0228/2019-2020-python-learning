## 臺南市獨居老人人數

```python
df=pd.read_csv("https://data.tainan.gov.tw/dataset/e56d1442-ea9f-4893-8cfe-b5a5f3b7fea1/resource/0cb5d6d8-2e47-42ca-9507-a26ae9002c40/download/1080529.csv")
```

### 區域獨居老人性別比

```python
df["區域獨居老人性別比(男/女)"]=df.男/df.女
```

```python
df
```

### 區域獨居老人性別比排序(高至低)

```python
df=df.sort_values(by=["區域獨居老人性別比(男/女)"],ascending=False)
```
