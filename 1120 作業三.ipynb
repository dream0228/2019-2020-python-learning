```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

## 3D圖練習

```python
from mpl_toolkits.mplot3d import Axes3D
```

##### 線形圖

```python
x=np.linspace(-10*np.pi,10*np.pi,1500)
y=np.cos(x)
z=np.sin(x)
fig = plt.figure()
ax = Axes3D(fig)
ax.scatter(x,y,z)
plt.show()
```

##### 子圖繪製

```python
fig = plt.figure() 

ax1=fig.add_subplot(1,2,1,projection='3d')

x=np.linspace(-10*np.pi,10*np.pi,1500)
y=np.cos(x)
z=np.sin(x)

ax1.plot(x,y,z)

ax2=fig.add_subplot(1,2,2,projection='3d')
X=np.arange(-4,4,0.25)
Y=np.arange(-4,4,0.25)
X,Y=np.meshgrid(X,Y)
Z=np.sqrt(X**4,Y**4)

ax2.plot_surface(X,Y,Z,cmap=plt.cm.summer)

plt.show()
```
