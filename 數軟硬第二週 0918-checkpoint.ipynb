## 條件判斷

```python
2>3                  # 寫程式時要小心不要蓋到關鍵字 ex.False,True
```

```python
8<9
```

```python
a=8
```

```python
a==9                 #判斷時要打兩個=否則a會變成9
```

```python
a!=9                 #!= 代表不等於
```

```python
a>=9                #>= 代表大於等於
```

```python
g=input("請輸入你的成績")                          # elif 條件判斷
g=float(g)                                        # else 剩餘質域判斷
if g<60:
   print("太混了啦")                               #會一路從上到下判斷，若有符合即停止
elif g>90:
    print("神人!")
elif g>70:
    print("還不錯啦")
else:
    print("那不就好棒棒")
```

```python
msg=input(">> ")
if("開心" in msg):
    print("真好")
else:
    print("拍拍")
```

## while 迴圈

#### 條件成立即執行，不成立則停止迴圈

```python
s=0
while s<10:
    print(s)
    s=s+1              #+1的用意再讓它不符合<10的定義，即停止迴圈
```

```python
msg="多拉A夢"
while(msg!="bye"):
    msg=input(">> ")
    print("拍拍!")
```
