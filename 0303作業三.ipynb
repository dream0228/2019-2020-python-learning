```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

### 畫圖練習

```python
x = np.linspace(0, 10, 1000)
y = np.sin(2*x)
z = np.cos(x**2)

plt.figure(figsize=(10,4))  #圖的大小
plt.plot(x,y,label="$sin(x)$",color="darkorange",linewidth=3)  #color=曲線顏色,linewidth=曲線寬度
plt.plot(x,z,"b--",label="$cos(x^2)$")
plt.xlabel("x")   #x軸名稱
plt.ylabel("y")      #y軸名稱
plt.title("figure 1")   #圖表名稱
plt.ylim(-2.2,2.2)    #y軸範圍
plt.legend()           #legend顯示圖示
plt.show()
```
