```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

## 3D圖練習

##### 圖一 

```python
from mpl_toolkits.mplot3d import Axes3D
```

```python
fig = plt.figure()
ax = Axes3D(fig)

X=np.arange(-3,3,0.5)
Y=np.arange(-3,3,0.5)
X,Y=np.meshgrid(X,Y)
Z=np.sqrt(X**2,Y**2)

ax.plot_surface(X,Y,Z,cmap=plt.cm.spring)

plt.show()
```

##### 圖二   (同時含兩種圖)

```python
fig = plt.figure()
ax = Axes3D(fig)

X=np.arange(-3,3,0.5)  #圖一
Y=np.arange(-3,3,0.5)
X,Y=np.meshgrid(X,Y)
Z=np.sqrt(X**2,Y**2)
ax.plot_surface(X,Y,Z,cmap=plt.cm.spring)


x1=np.linspace(-2*np.pi,2*np.pi,1000)  #圖二
y1=np.sin(x1)
ax.plot(x1,y1,zs=0,c='g')

plt.show()
```
