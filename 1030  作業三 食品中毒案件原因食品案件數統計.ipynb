## 食品中毒案件原因食品案件數統計

```python
df=pd.read_csv("https://quality.data.gov.tw/dq_download_csv.php?nid=9837&md5_url=47592a897e044444ca803802e04eda72")
```

```python
df
```

```python
df["總合計"]=df.loc[:,"原因食品判明合計":"原因食品不明合計"].sum(axis=1)
```

```python
df
```

### 中毒事件最多排序

```python
df=df.sort_values(by=["總合計"],ascending=False)
```

```python
df
```

## 已知原因食物中，複合調理食品佔比

```python
df1=df[["年度","原因食品判明合計","複合調理食品"]]
```

```python
df1
```

```python
df["複合調理食品引起判明事件比例"]=df1.複合調理食品/df1.原因食品判明合計
```
