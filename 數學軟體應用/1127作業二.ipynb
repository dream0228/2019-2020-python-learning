### 作業二   3D圖練習

```python
from mpl_toolkits.mplot3d import Axes3D

from matplotlib import cm
```

```python
fig = plt.figure()
ax = fig.gca(projection='3d')

x = np.arange(-100, 100,0.6)
y = np.arange(-10, 10,0.5)
X, Y = np.meshgrid(x, y)

Z = np.add(-np.power(X, 4), np.power(Y, 4))

ax.set_xlabel("X")
ax.set_ylabel("Y")
ax.plot_wireframe(X, Y, Z, alpha=0.2)
ax.contour(X, Y, Z, cmap=cm.gist_rainbow, linewidths=2)

plt.show()
```
