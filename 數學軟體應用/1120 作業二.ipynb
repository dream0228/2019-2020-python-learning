```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

## contour練習-數據畫圖

```python
z_list = []
x_list = []
y_list = []
for x in range(3):
    for y in range(3):
        z = x**3+4*y**3
        z_list.append(z)
        x_list.append(x)
        y_list.append(y)
x,y,z = x_list,y_list,z_list
x = np.array(x)
x.shape = (3,3)
x = np.mat(x).T
y = np.array(y)   
y.shape = (3,3)
y = np.mat(y).T
z = np.array(z)
z.shape = (3,3)
z = np.mat(z).T
plt.figure(figsize=(10,8))
plt.contourf(x,y,z)
plt.contour(x,y,z)
```
