```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

## contour畫圖

```python
step = 0.005    
x = np.arange(-10,10,step)
y = np.arange(-10,10,step)
```

```python
X,Y = np.meshgrid(x,y)
```

```python
Z = 7*X**2+13*Y**2-77
```

```python
plt.figure(figsize=(10,15))  
plt.contourf(X,Y,Z) 
```

```python
contour = plt.contour(X,Y,Z,[100,200,300,400],colors='g')     
plt.clabel(contour,fontsize=10,colors=('b','r','b','r'))     
```
